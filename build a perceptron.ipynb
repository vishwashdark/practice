```python
#creating a perceptron 
import math 
import numpy as np
def g_sigmoid(x):
    g=1/(1+math.exp(-x))
    return g
def weights(x,w):
    x=np.array(x)
    w=np.array(w)
    if len(x)!=len(w):
        return False
    else:
        p=np.dot(w,x)
        return p
def g_bipolar(p):
    return (2/(1+math.exp(p)))-1
```

```python
def rnn_step(word, hidden_state):
    # Dummy implementation of an RNN step
    # Update hidden_state and generate prediction based on the current word
    updated_hidden_state = [h + 1 for h in hidden_state]  # Example: increment hidden state
    prediction = f"Next_{word}"  # Example: generate dummy prediction
    return prediction, updated_hidden_state

def rnn():
    hidden_state = [0, 0, 0, 0]  # Initial hidden state
    sentence = ["I", "Love", "Coding", "Maths"]  # Input sequence
    for word in sentence:
        p, hidden_state = rnn_step(word, hidden_state)
        print(f"Word: {word}, Prediction: {p}, Hidden State: {hidden_state}")
    next_word = p  # Final prediction
    return next_word

rnn()
```
